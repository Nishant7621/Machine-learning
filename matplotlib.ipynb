# 3D surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y) # x and y are 1D arrays. Typically, x represents the horizontal (x-axis) values, and y represents the vertical (y-axis) values.
#xx and yy are 2D arrays, where
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)


```

```python
fig= plt.figure(figsize=(12,8))
ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig= plt.figure(figsize=(12,8))
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        x_val = xx[i, j]
        y_val = yy[i, j]
        z_val = z[i, j]
        print(f"Coordinate (x, y, z) at index ({i}, {j}): ({x_val}, {y_val}, {z_val})")
```

```python
z= np.sin(xx) + np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
