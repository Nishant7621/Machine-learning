```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data=pd.read_csv(url,parse_dates=["Month"],index_col="Month")
data.head()
```

```python
data.head()
```

```python
plt.figure(figsize=(12,8))
plt.plot(data,label="monthly sales")
plt.xlabel("time")
plt.ylabel("passengers count")

plt.legend()
plt.show()
```

```python
import matplotlib.dates as mdates

plt.figure(figsize=(12,8))
plt.plot(data["Passengers"],label="number of passengers")
plt.xlabel("time")
plt.ylabel("passengers")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=10))
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

add_seasonal = seasonal_decompose(data["Passengers"],model="additive")
add_seasonal.plot()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

add_seasonal = seasonal_decompose(data["Passengers"],model="multiplicative")
add_seasonal.plot()
plt.show()
```
