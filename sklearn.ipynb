```python
df['smoker']= df['smoker'].map({'Yes':1, 'No':0})
df.head(5)
```

```python
X=df[['total_bill','smoker','size']]
y=df[['tip']]
model = LinearRegression()
model.fit(X,y)
print(model.predict([[35.527,1,2]]))
print(model.predict([[24.59,0,4]]))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
X=df[['total_bill','size']]
y=df[['tip']]
model=DecisionTreeRegressor()
model.fit(X,y)
model.predict([[50.527,2]])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8)
model=LinearRegression()

model.fit(X_train,y_train)
#metric to evaluate yhe model
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
y_pred = model.predict(X_test)

print(f"Mean Squarred Error: {mean_squared_error(y_test,y_pred)}")
print(f"R2_score: {r2_score(y_test,y_pred)}")
print(f"Mean Squarred Error: {mean_absolute_error(y_test,y_pred)}")

```
